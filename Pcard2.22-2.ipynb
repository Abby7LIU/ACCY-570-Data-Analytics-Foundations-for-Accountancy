```python
def convert_date(date_str):
    '''
    date format: https://www.tutorialspoint.com/python/time_strptime.htm
    existing formats:
    28-Jun-1989
    18.07.2012 00:00:00
    3/25/2013
    3/4/2014 0:00
    '''
    try:
        #28-Jun-1989
        dt = datetime.strptime(date_str, '%d-%b-%Y')
        return dt#.strftime('%Y-%m-%d')
    except:
        pass
    try:
        #3/25/18
        dt = datetime.strptime(date_str, '%m/%d/%y')
        return dt#.strftime('%Y-%m-%d')
    except:
        pass
    try:
        #28-Jun-89
        dt = datetime.strptime(date_str, '%d-%b-%y')
        return dt#.strftime('%Y-%m-%d')
    except:
        pass
    try:
        #18.07.2012 00:00:00
        dt = datetime.strptime(date_str, '%d.%m.%Y %H:%M:%S')
        return dt#.strftime('%Y-%m-%d')
    except:
        pass
    try:
        #2/18/1980
        dt = datetime.strptime(date_str, '%m/%d/%Y')
        return dt#.strftime('%Y-%m-%d')
    except:
        pass
    try:
        #2/18/1980 00:00
        dt = datetime.strptime(date_str, '%m/%d/%Y %H:%M')
        return dt#.strftime('%Y-%m-%d')
    except:
        pass
    print (date_str)
```

```python
pcard['Transaction_Date'] = pcard['Transaction_Date'].apply(lambda x:convert_date(x))
pcard['Posted_Date'] = pcard['Posted_Date'].apply(lambda x:convert_date(x))
```
