```python
loan = pd.read_csv('loan.csv', low_memory=False)
loan['repaid'] = loan['total_pymnt_inv'] > loan['funded_amnt_inv']
loan_a =loan[['loan_amnt', 'term', 'int_rate', 'grade', 
             'emp_length', 'home_ownership', 'annual_inc', 
             'verification_status', 'purpose', 'dti', 'revol_util', 'total_acc', 
              'total_rec_prncp', 'total_rec_int',
              'funded_amnt','installment','sub_grade',
             'open_acc','repaid']]
loan_a = loan_a.dropna()
# Generate categorical features
categorical = ['grade', 'term', 'emp_length','home_ownership', 
               'verification_status', 'purpose','sub_grade']
cat_data = pd.get_dummies(loan_a[categorical])

# Extract numerical features
numerical = ['loan_amnt', 'int_rate','annual_inc',
            'dti', 'revol_util', 'total_acc', 
            'total_rec_prncp', 
             'total_rec_int','funded_amnt',
             'installment',
             'open_acc']
num_data = loan_a[numerical]

# Combine categorical and numerical data into features
x = pd.concat([num_data, cat_data], axis=1)
x.sample(5)
```

In order to measure the relationship between categorical feature and numerical feature, we divided the numerical data into several intervals, and then calculated the probability of of repayment in each interval. After that, we drawed plots to show the trend between repaid and numerical features.

```python
import numpy as np
feature_dti = np.asarray(loan_a['dti'])
feature_repaid = np.asarray(loan_a['repaid'])
upBound = feature_dti.max()
lowBound = feature_dti.min()
[upBound, lowBound] # [34.99 0.0]

interval = 1

iterations = int(upBound/interval) + 1

result = np.zeros(iterations)
for i in range(iterations):
    sampleNum = 0
    positiveSampleNum = 0
    
    for j in range(len(feature_dti)):
        if feature_dti[j] > i * interval and feature_dti[j] < ((i+1) * interval) :
            sampleNum += 1
            
            if feature_repaid[j] == True:
                positiveSampleNum += 1
    
    if sampleNum != 0:
        result[i] = (positiveSampleNum * 1.0) / sampleNum

axis = np.linspace(interval/2, (iterations) * interval - interval/2, iterations)
print (axis)
import matplotlib.pyplot as plt
plt.plot(axis, result)
plt.ylabel('Probability of Repayment', fontsize=18)
plt.xlabel('DTI', fontsize=18)

plt.title('DTI vs Repaid', ha='center', fontsize=20)
```

Based on the plot we can see a downward trend between DTI and repaid ratio. With lower DTI, borrowers are more likely to repay their loans.

For other visualization to present the relationship between the features and repaid, we have chosen gains chart and lift chart. From the gains chart, we can see that Random Forest Classifier basically overlaps with the baseline. Gradient Boosting is slightly better than Random Forest.

```python
# Write a function to calculate gains  

def compute_gain(mdl, d_test, l_test):

    
    if hasattr(mdl, 'decision_function'):
        prbas = mdl.decision_function(d_test)
        pos_score = prbas

    else:
        prbas = mdl.predict_proba(d_test)
        pos_score = prbas[:,1]

    clm = pd.get_dummies(l_test).as_matrix()
    pos_lbl = clm[:,1]
    n_pos_lbl = np.sum(pos_lbl)
    idx = np.argsort(pos_score) 
    sort_pos = pos_lbl[idx[::-1]]
    cum_sum_pos = np.cumsum(sort_pos)
    lift = cum_sum_pos/n_pos_lbl
    num_test = l_test.shape[0]
    tst_pcnt = np.arange(1,num_test + 1, 1) /num_test
    
    return tst_pcnt, lift
```

```python
# Run Logistic Regression as basedline for gain chart
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=23)
lr = lr.fit(features_train, repaid_train)
```

```python
# Compute gains for different estimators
xgbtc, gain_gbtc = compute_gain(gbtc, features_test, repaid_test)
xrfc, gain_rfc = compute_gain(rfc, features_test, repaid_test)
xlr, gain_lr = compute_gain(lr, features_test, repaid_test)

# Make the plot
fig, ax = plt.subplots(figsize=(8, 8))

ax.plot(xlr, xlr, alpha = 0.5, linestyle='-',
        label=f'Baseline')

ax.plot(xgbtc, gain_gbtc, alpha = 0.75, linestyle='-.',
        label=f'GBTC Gain')

ax.plot(xrfc, gain_rfc, alpha = 0.75, linestyle=':',
        label=f'RFC Gain')


ax.set_title('Gain Chart', fontsize=18)
ax.set_xlabel('Fraction (test data)', fontsize=16)
ax.set_ylabel('Gain', fontsize=16)
ax.set_xlim(-0.05, 1.05)
ax.set_ylim(-0.05, 1.05)
ax.set_aspect('equal')
ax.legend(loc=4, fontsize=16)
sns.despine(offset=5, trim=True)
plt.show()
```

From the lift chart, it is clearer that Gradient Boosting is a better model than Random Forest because the lift is higher. 

```python
# Compute lift by normalzing gains to baseline
lift_gbtc = gain_gbtc / xgbtc
lift_rfc = gain_rfc / xrfc

# Make the plot
fig, ax = plt.subplots(figsize=(8, 8))

ax.plot(xgbtc, lift_gbtc, alpha = 0.75, linestyle='-.',
        label=f'GBTC Lift')

ax.plot(xrfc, lift_rfc, alpha = 0.75, linestyle=':',
        label=f'RFC Lift')

ax.hlines(1.0, 0.0, 1.0, alpha = 0.5, linestyle='--',
        label=f'Baseline')

# Decorate plot appropriately
ax.set_title('Lift Chart', fontsize=18)
ax.set_xlabel('Fraction (test data)', fontsize=16)
ax.set_ylabel('Lift', fontsize=16)
ax.set_xlim(0, 1)
ax.set_ylim(0.95, 1.2)
ax.legend(loc=1, fontsize=16)
sns.despine(offset=5, trim=True)
plt.show()
```

## Machine Learning

### Data Preparation

```python
dfc = loan[['int_rate','installment','revol_util','term','annual_inc','purpose','repaid']]
dfd = loan['return']
    
tm2 = pd.concat([dfc, dfd], axis=1)
tm3=tm2.dropna()


dfe = tm3[['int_rate','installment','revol_util','term','annual_inc','purpose','repaid']]
dff = tm3['return']
```

#### Normalize

```python
from sklearn.preprocessing import StandardScaler 

ss = StandardScaler()

int_rate = ss.fit_transform(dfe['int_rate'])
installment = ss.fit_transform(dfe['installment'])
revol_util = ss.fit_transform(dfe['revol_util'])
annual_inc= ss.fit_transform(dfe['annual_inc'])



dfe['int_rate'] = int_rate
dfe['installment'] = installment
dfe['revol_util'] = revol_util
dfe['annual_inc'] = annual_inc
```

#### Splitting the data

```python
X_train, X_test, y_train, y_test \
= train_test_split(dfe, dff, test_size= 0.25, random_state = 0)
```

### KNeighbors Regressor

```python
from sklearn.neighbors import KNeighborsRegressor as KNR
KNN = KNR(n_neighbors = 1)
KNN.fit(X_train, y_train)
    
prediction_KNN = KNN.predict(X_test)

score = KNN.score(X_test, y_test)
print("accuracy = {0:3.1f} %.".format(100.0 * score))

ax_KNN = sns.residplot(y_test, prediction_KNN,)
ax_KNN.set_ylabel("Residual", fontsize=14)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression as LR
LR = LR(fit_intercept=False)
LR.fit(X_train, y_train)
    
prediction_LR = LR.predict(X_test)

score = LR.score(X_test, y_test)
print("accuracy = {0:3.1f} %.".format(100.0 * score))

ax_LR = sns.residplot(y_test, prediction_LR, color = 'green')
ax_LR.set_ylabel("Residual", fontsize=14)
```

### RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor as RFR
rfr = RFR(random_state=0)
rfr.fit(X_train, y_train)
    
prediction_rfr = rfr.predict(X_test)

score = rfr.score(X_test, y_test)
print("accuracy = {0:3.1f} %.".format(100.0 * score))

ax_rfr = sns.residplot(y_test, prediction_rfr, color = 'orange')
ax_rfr.set_ylabel("Residual", fontsize=14)
```

__Analysis of residual plot__    
The residual plot shows the difference between predicted value and observed value. From the plot above, we can see a 'line' vertical to the x-axis when 'return' is about '0', suggesting that the models tend to predict more returns when the true return is a little higher than 0, and less returns when true return is a little less than 0.The points are dispersed around the horizontal axix, and the number of points above the x-axis is similar to the the mumber of the points below the x-axis. Thus, all regession models are relatively appropriate for the data.
